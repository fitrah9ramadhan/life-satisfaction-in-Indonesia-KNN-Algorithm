### Collecting Data IFLS 5

```python
import pandas as pd
import numpy as np
```

```python
#hr6 = self_own_assets

df0 = pd.read_stata("/IFLS/IFLS 5/ALL DTA/b3a_hr1.dta")
df0 = df0[["hr06", "pidlink","hhid14_9"]].dropna()
df0 = df0.rename(columns = {"hr06":"self_own_assets"})
len(df0)
```

```python
df0["self_own_assets"].unique()
```

```python
self_own_assets_dict = {"1:Yes":1, "3:No":3}
```

```python
df0["self_own_assets"] = df0["self_own_assets"].map(self_own_assets_dict)
```

```python
#sw00 = satisfaction
#sw01 = poor_level
#sw02 = poor_level_5y_ago
#sw03 = poor_level_5y_later

df1 = pd.read_stata("/IFLS/IFLS 5/ALL DTA/b3a_sw.dta")
df1 = df1[["sw00", "sw01", "sw02", "sw03", "pidlink", "hhid14_9"]].dropna()
df1 = df1.rename(columns = {"sw00":"life_satisfaction" ,"sw01":"poor_level" ,"sw02":"poor_level_5y_ago" ,"sw03":"poor_level_5y_later"})
len(df1)
```

```python
print(df1["life_satisfaction"].unique())

print(df1["poor_level"].unique())
print(df1["poor_level_5y_ago"].unique())
print(df1["poor_level_5y_later"].unique())
```

```python
satisfaction_dict = {"5:Not at all satisfied": 5, "3:Somewhat satisfied": 3,"4:Not very satisfied":4 ,"2:Very satisfied":2 ,"1:Completely satisfied":1 ,"9:MISSING": np.nan}
poor_level_5y_ago = {"1:Poorest":1, "6:Richest":6, "8:Don't Know":np.nan, "9:MISSING":np.nan}
poor_level_dict = {"1:Poorest":1, "6:Richest":6, "8:Don't Know":np.nan, "9:MISSING":np.nan}
poor_level_5y_later_dict = {"1:Poorest":1, "6:Richest":6, "8:Don't Know":np.nan, "9:MISSING":np.nan}
```

```python
df1["life_satisfaction"] = df1["life_satisfaction"].map(satisfaction_dict)
df1["poor_level"] = df1["poor_level"].map(poor_level_dict)
df1["poor_level_5y_ago"] = df1["poor_level_5y_ago"].map(poor_level_5y_ago)
df1["poor_level_5y_later"] = df1["poor_level_5y_later"].map(poor_level_5y_later_dict)


df1.dropna(inplace = True)
df1
```

```python
#tk16c1 = job_satisfaction

df2 = pd.read_stata("/IFLS/IFLS 5/ALL DTA/b3a_tk1.dta")
df2 = df2[["tk16c1", "pidlink", "hhid14_9"]]
df2 = df2.rename(columns = {"tk16c1":"job_satisfaction"})
len(df2)
```

```python
df2["job_satisfaction"].unique()
```

```python
job_satisfaction_dict = {"2:Satisfied":2, "3:Unsatisfied":3 ,"1:Very satisfied":1 ,"4:Very unsatisfied":4 ,"9:Missing":np.nan}
```

```python
df2["job_satisfaction"] = df2["job_satisfaction"].map(job_satisfaction_dict)
df2.dropna(inplace = True)
df2
```

```python
# tk21a = working hours

df3 = pd.read_stata("/IFLS/IFLS 5/ALL DTA/b3a_tk2.dta")
df3 = df3[['tk21a', "pidlink", "hhid14_9"]].dropna()
df3 = df3.rename(columns = {"tk21a":"working hours"})

df3
```

```python
df_merge1 = df0.merge(df1, on = "pidlink")
df_merge2 = df_merge1.merge(df2, on = "pidlink")
df_merge3 = df_merge2.merge(df3, on = "pidlink")


df_merge3
```

```python
df_merge3 = df_merge3[["self_own_assets", "life_satisfaction","poor_level","poor_level_5y_ago","poor_level_5y_later","job_satisfaction","working hours"]]
df_merge3
```

```python
data_fix = df_merge3[["self_own_assets", "poor_level", "poor_level_5y_ago", "poor_level_5y_later", "job_satisfaction", "working hours", "life_satisfaction"]]
data_fix.to_csv("data_fix.csv")
```
